```python
library("FactoMineR")
library("factoextra")
library("psych")
library("ggplot2")
library("corrplot")
```

```python
train_full<-read.csv(file='data/train.csv')
```

```python
nums<- unlist(lapply(train_full, is.numeric))
var.pca<-PCA(train_full[,nums])
```

```python
fviz_pca_var(var.pca, col.var="red")
```

```python
library("corrplot")
train_full_nums <- train_full[,nums]
matcor <- cor(train_full_nums[,c(-3,-9,-26)])
corrplot(matcor)
```
